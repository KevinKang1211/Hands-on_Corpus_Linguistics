## 資料擷取

### 網頁資料 (爬蟲)

```
import requests, requests_cache, time
from bs4 import BeautifulSoup
requests_cache.install_cache()
```

```
resp = requests.get("https://tw.news.yahoo.com/")
assert resp.status_code == 200
```

```
html = BeautifulSoup(resp.text, 'html5lib')
```

```
with open("test.txt", "w", encoding="UTF-8") as fout:
  fout.write(resp.text)
```

```
news_pages = []
for aelem in html.select("div#mrt-node-Col1-1-Hero ul ul a")[:3]:    

    resp_page = requests.get(aelem["href"])
    from_cache = resp_page.from_cache
    print("Retrieving ", aelem.text, "...", end='')
    print("[cache:", from_cache, "]")

    if resp.status_code != 200:
        print("Status code: ", resp.status_code)
        print(resp.text)
        continue

    page_html = BeautifulSoup(resp_page.text, 'html5lib')
    title = page_html.select_one("header.caas-title-wrapper").text
    content = page_html.select_one("div.caas-content-wrapper").text
    news_pages.append((title, content))

    
    if not from_cache:
        time.sleep(0.5)
```

## 文字編碼

### 單一字元編碼

```
def print_code(ch, encoding="UTF-16BE"):
    ch_int = int.from_bytes(ch.encode(encoding), 'big')
    return f"{ch_int} (0x{ch_int:X})"
```

```
print_code("A")
```

```
# cf. https://www.unicode.org/cgi-bin/GetUnihanData.pl?codepoint=%E8%AA%9E
print_code("語", "UTF-16BE"), print_code("語", "Big5"), print_code("語", "UTF-8")
```

```
# cf. https://emojipedia.org/face-with-tears-of-joy/
print_code("😂", "UTF-32BE"), print_code("😂", "utf-8")
```

### 檔案編碼與偵測

```
import chardet
```

```
print("Big5 文字檔案")
with open("data_1.2/text.sample.big5.txt", "rb") as fin:
    bin_data = fin.read()    
    print("檔案內容: ", bin_data[:16])
    print("編碼偵測: ", chardet.detect(bin_data))
print()
print("UTF-8 文字檔案")
with open("data_1.2/text.sample.utf8.txt", "rb") as fin:
    bin_data = fin.read()    
    print("檔案內容: ", bin_data[:16])
    print("編碼偵測: ", chardet.detect(bin_data))
```

```
with open("data_1.2/text.sample.utf8.txt", "r", encoding="UTF-8") as fin:
    print(fin.read())
```
