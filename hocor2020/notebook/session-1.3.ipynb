## 資料整理

* 把資料整理成一個List of Dict，每個Dict是一篇「文章」。文章有三個「屬性」：查詢詞(`query`)、標題(`title`)、原始材料(`rawtext`)

```
corpus = []
for key, data_item in data.items():
    article = {}
    article["query"] = key
    article["title"] = data_item[0]
    article["rawtext"] = data_item[1]
    corpus.append(article)
```

```
corpus[:3]
```

### 簡單的描述性資料

```
n_char_raw = sum(len(x["rawtext"]) for x in corpus)
print("文章個數: ", len(corpus))
print("文章總長度: ", n_char_raw)
print("平均文章長度: ", round(n_char/len(corpus), 2))
```

```
zhon.hanzi.characters
```

```
import re
import zhon.hanzi
zh_pat = f"[{zhon.hanzi.characters}|{zhon.hanzi.punctuation}]"
for article in corpus:
    zhstream = "".join(re.findall(zh_pat, article["rawtext"]))
    article["preproc"] = re.findall(zhon.hanzi.sentence, zhstream)
```

```
corpus[:2]
```

```
n_char_preproc = sum([sum(len(y) for y in x["preproc"]) for x in corpus])
print("文章原始總長度: ", n_char_raw)
print("文章前處理後總長度: ", n_char_preproc)
print("前處理保留比例: ", round(n_char_preproc/n_char_raw, 2))
```

## 檔案儲存

### 純文字

```
with open("corpus-wiki.txt", "w", encoding="UTF-8") as fout:
    for article in corpus:
        fout.write("\n".join(article["preproc"]))
        fout.write("\n")
```

### 逗號分隔檔

```
import csv
with open("corpus-wiki.csv", "w", encoding="UTF-8") as fout:
    writer = csv.writer(fout)
    writer.writerow(["query", "title", "preproc"])
    for article in corpus:
        writer.writerow([article["query"], article["title"], "\n".join(article["preproc"])])
```

### JSON格式

```
import json
with open("corpus-wiki.json", "w", encoding="UTF-8") as fout:
    json.dump(corpus, fout, ensure_ascii=False, indent=2)
```
